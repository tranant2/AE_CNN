## Calculate error and sd

```python
diff = np.zeros((4, losses[0].shape[0]))
for i in range(4):
    diff[i] = test_targets[i]-losses[i].cpu()  # If positive error, model tends to underestimate
```

```python
for i in range(4):
    u = np.mean(diff[i])
    sd = np.std(diff[i])
    print(f"For the {i} loss location, mean:{u}. std:{sd}")
```

```python
fig = plt.figure(figsize=(16,4))
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.tight_layout()
    plt.hist(np.abs(diff[i]), bins=100)
    plt.xlabel('True-Predicted', fontsize=14)
    plt.title(f"Loss: {i}", fontsize=18)
    plt.xlim(0,1000)
    plt.ylim(0,150)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    u = np.mean(np.abs(diff[i]))
    sd = np.std(np.abs(diff[i]))
    plt.text(50, 135, "mean: {:.2f}".format(u), fontsize=14)
    plt.text(50, 125, "std: {:.2f}".format(sd), fontsize=14)
fig.savefig(f"/mnt/home/tranant2/Desktop/MachineLearning/TRACK/CNN9/Histogram_Final.png", dpi = 300)
```
